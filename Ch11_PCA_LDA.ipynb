##### 11.4.1 필요한 패키치 Import

```python
# Load libraries
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

##### 11.4.2 원본 데이터 확인

```python
# Load the data
data = datasets.load_breast_cancer()
```

```python
# Check the data
print("Feature Names:", data.feature_names)
print("Number of Features:", len(data.feature_names))
```

##### 11.4.3 데이터 셋 준비

```python
x = data.data[:, :2]
y = data.target
target_names = data.target_names
```

```python
plt.figure(figsize = (10,10))
colors = ['red', 'blue']
for color, i, target_name in zip(colors, [0,1], target_names):
    plt.scatter(x[y == i, 0], x[y == i, 1], color = color, label = target_name, s = 5)

plt.legend()
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.show()
```

##### 11.4.3 PCA

```python
# Load the Data
x = data.data
y = data.target
target_names = data.target_names
```

```python
# Reduce the Demensionality using PCA
pca = PCA(n_components = 2)
x_p = pca.fit(x).transform(x)
print("가장 큰 주성분 두 개에 대한 분산 {}".format(str(pca.explained_variance_ratio_)))
```

```python
plt.figure(figsize = (10,10))
colors = ["red", "blue"]

for color, i, target_name in zip(colors, [0,1], target_names):
    plt.scatter(x_p[y == i, 0], x_p[y == i, 1], color = color, label = target_name, s = 5)
    
plt.legend()
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

##### 11.4.4 LDA

```python
# Load the Data
x = data.data
y = data.target
target_names = data.target_names
```

```python
# Reduce the Demensionality using LDA
lda = LinearDiscriminantAnalysis(solver = 'eigen', n_components = 2)
x_l = lda.fit(x, y).transform(x)
```

```python
plt.figure(figsize = (10,10))
colors = ["red", "blue"]

for color, i, target_name in zip(colors, [0,1], target_names):
    plt.scatter(x_l[y == i, 0], x_l[y == i, 1], color = color, label = target_name, s = 5)
    
plt.legend()
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
